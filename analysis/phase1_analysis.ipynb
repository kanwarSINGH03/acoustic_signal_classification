```python
import plotly.express as px
from sklearn.decomposition import PCA
import scipy.io as sio
import pandas as pd

```

```python
phase1_data = sio.loadmat('../data/mine_impact_data_2019.mat')
print(phase1_data.keys())
```

```python

df_x  = pd.DataFrame(phase1_data["x"].T)
df_y  = pd.DataFrame(phase1_data["y"].T, columns=["y"])
df_mine = pd.DataFrame(phase1_data["mine"].T, columns=["mine"])

df = pd.concat([df_x, df_y, df_mine], axis=1, join="inner")

df = df.dropna()



display(df)
```

```python
X = df.iloc[:, :-2]

pca = PCA(n_components=100)

components = pca.fit_transform(X)

fig = px.scatter(components, x=0, y=1, color=df['y'], hover_name=df['mine'],)
fig.show()
```
