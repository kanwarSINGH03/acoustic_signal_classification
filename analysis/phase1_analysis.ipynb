```python
import plotly.express as px
from sklearn.decomposition import PCA
import scipy.io as sio
import pandas as pd
import numpy as np
```

```python
phase1_data = sio.loadmat('../data/mine_impact_data_2019.mat')
print(phase1_data.keys())
```

```python

df_x  = pd.DataFrame(phase1_data["x"].T)
df_y  = pd.DataFrame(phase1_data["y"].T, columns=["y"])
df_mine = pd.DataFrame(phase1_data["mine"].T, columns=["mine"])

df = pd.concat([df_x, df_y, df_mine], axis=1, join="inner")

df = df.dropna()



display(df)
```

```python
X = df.iloc[:, :-2]

pca = PCA(n_components=2)

components = pca.fit_transform(X)

new_axis = phase1_data["y"].T.copy().astype(float)
new_axis[new_axis == 0] = -1

df_components = pd.DataFrame(components, columns=["PC1", "PC2"])
df_new_axis = pd.DataFrame( new_axis, columns=["z"])

df_kernel = pd.concat([df_components, df_new_axis], axis=1, join="inner")

one_count = np.sum(df_kernel["z"] == 1)
print("one_count", one_count)
zero_count = np.sum(df_kernel["z"] == -1)
print("zero_count", zero_count)

fig = px.scatter_3d(df_kernel, x="PC1", y="PC2", z="z", color=df['y'], hover_name=df['mine'],)
fig.show()
```
