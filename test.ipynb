```python
import pandas as pd
import numpy as np
import scipy.io as sio
```

```python
phase1_data = sio.loadmat('./data/mine_impact_data_2019.mat')
samples  = pd.DataFrame(phase1_data["x"].T)
labels  = pd.DataFrame(phase1_data["y"].T, columns=["y"])
mines = pd.DataFrame(phase1_data["mine"].T, columns=["mine"])

df = pd.concat([samples, labels, mines], axis=1, join="inner")

df = df.dropna()
df = df.reset_index(drop=True)
display(df)
```
